```python
#Storing rows in a .txt files, containing one of our contributor's travel records. This exported using Google Maps API.
#The object contains the following parameters:
#travel method (e.g. walking, passenger vehicle ect.), travel duration, travel distance, travel method rating (based on CO2 emissions)
```

```python
import json
f = open("2017_NOVEMBER.txt")
fc = f.read()
obj = json.loads(fc)
obj = obj["timelineObjects"]
class CO2Trip:
    def __init__(self, b,d,dist,act_type):
        self.begin_time = b
        self.duration = d
        self.distance = dist
        self.action = act_type
        
        self.green_type = ""
        if self.action in ["WALKING", "CYCLING"]:
            self.green_type = "GREEN"
        elif self.action == "IN_BUS":
            self.green_type = "BUS"
        elif self.action in ["IN_PASSENGER_VEHICLE", "MOTORCYCLING"]:
            self.green_type = "UBER"
        elif self.action == "IN_SUBWAY":
            self.green_type = "SUBWAY"
        else:
            print(self.begin_time, self.action)
            self.green_type = self.action
    def __repr__(self):
        return "(Date: {}, Duration: {}, Distance: {}, {}, Green? {})".format(self.begin_time, self.duration,self.distance, self.action,self.green_type)
    def __str__(self):
        return self.__repr__()
from dateutil import parser
activities = []
interest_field = ["duration", "distance", "activityType", "confidence", "distanceMeters"]
for v in obj:
    if 'activitySegment' in v:
        v = v["activitySegment"]
        d = v["duration"]
        bt = d["startTimestamp"]
        bt = parser.isoparse(bt).date()
        d = (parser.isoparse(d["endTimestamp"]) - parser.isoparse(d["startTimestamp"])).total_seconds()
        dis = -1
        if "distance" in v:
            dis = v["distance"]
        elif "distanceMeters" in v:
            dis = v["distanceMeters"]
        elif "waypointPath" in v and "distanceMeters" in v["waypointPath"]:
            dis = v["waypointPath"]["distanceMeters"]
        elif "simplifiedRawPath" in v and "distanceMeters" in v["simplifiedRawPath"]:
            dis = v["simplifiedRawPath"]["distanceMeters"]
        else:
            print(json.dumps(v,indent="  "))

        a = v["activityType"]
        
        # print(dis/d)
        if a in ["UNKNOWN_ACTIVITY_TYPE", "FLYING"]:
            continue
            
        activities.append(CO2Trip(bt,d,dis,a))
len(activities)
for a in activities:
  print(a)
  
```

```python
#We want WalkingTrees to provide a summary of the user's travel activity per day
#Specifically, we want our user to see what transportation methods they used and over what distance
#Therefore, for each Day object, we are interested in its 'transportation method' and 'distance travelled' values
```

```python
trips = pd.DataFrame(columns = ["Date", "GREEN", "BUS", "UBER", "SUBWAY"])
trips.head
```

```python
for i in range(len(activities)):
  day = 0
  green = 0
  bus = 0
  uber = 0
  subway = 0
  trips.at[i, 'Date'] = activities[i].begin_time
  if activities[i].green_type == "GREEN":
    trips.at[i, 'GREEN'] = activities[i].distance
  if activities[i].green_type == "BUS":
    trips.at[i, 'BUS'] = activities[i].distance
  if activities[i].green_type == "UBER":
    trips.at[i, 'UBER'] = activities[i].distance
  if activities[i].green_type == "SUBWAY":
    trips.at[i, 'SUBWAY'] = activities[i].distance 

trips.head
```

```python
trips1 = pd.DataFrame(columns = ["Date", "GREEN", "BUS", "UBER", "SUBWAY"])
for i in range(len(activities1)):
  day = 0
  green = 0
  bus = 0
  uber = 0
  subway = 0
  trips1.at[i, 'Date'] = activities1[i].begin_time
  if activities1[i].green_type == "GREEN":
    trips1.at[i, 'GREEN'] = activities1[i].distance
  if activities1[i].green_type == "BUS":
    trips1.at[i, 'BUS'] = activities1[i].distance
  if activities1[i].green_type == "UBER":
    trips1.at[i, 'UBER'] = activities1[i].distance
  if activities1[i].green_type == "SUBWAY":
    trips1.at[i, 'SUBWAY'] = activities1[i].distance 

trips1.head
```

```python
#To inform the user's change in transportation habits over time, we need short-term as well as long-term data
#Therefore, I calculate the user's aggregate travel pattern (in terms of transportation methods-  
#and distance) over the past week, the past month, and the past year.
```

```python
trip_day = trips.groupby("Date").sum()
trip_day
```

```python
trip_day1 = trips1.groupby("Date").sum()
trip_day1
```

```python
trips_month = trip_day1.sum()
trips_month
```

```python
trip_all = pd.concat([trip_day, trip_day1])
trip_all
```

```python
trips_year = trip_all.sum()
trips_year
```

```python
trip_all.to_csv("day.csv")
trips_year.to_csv("year.csv")
trips_month.to_csv("month.csv")
```
